```python
from numpy.fft import fft, ifft, fftshift, ifftshift, fftfreq
import numpy as np
```

```python


def manakovSSF(Ex, Ey, hz, Lspan, Ltotal, alpha, gamma, D, Fc, Fs):
    
    c = 299792458
    c_kms = c/1e3
    λ = c/Fc
    α  = alpha/(10*np.log10(np.exp(1)))
    β2 = -(D*λ**2)/(2*np.pi*c/1e3)
    γ = gamma
    
    l = len(Ex)
    
    Nfft = l;

#     if (Nfft/2)%1 == 0:
#       #  ω = 2*pi*[(0:Nfft/2-1),(-Nfft/2:-1)]'/(Nfft/Fs);
#         ω = 2*pi*np.append(np.range(0,Nfft/2), np.range(-Nfft/2,0))/(Nfft/Fs)
#     else
#         ω = 2*pi*[(0:Nfft/2-.5),(-Nfft/2:-1)]'/(Nfft/Fs);
   
  #  ω = 2*np.pi*np.arange(-Nfft/2, Nfft/2, 1)*(Fs/Nfft)
    ω = 2*np.pi*Fs*fftfreq(Nfft)
    
    Nspans = int(np.floor(Ltotal/Lspan))
    Nsteps = int(np.floor(Lspan/hz))

    Ex = fft(Ex) #Pol. X spectrum
    Ey = fft(Ey) #Pol. Y spectrum
    
    for spanN in range(1, Nspans+1):
        for stepN in range(1, Nsteps+1):
            
            # First linear step:
            Ex = Ex*np.exp(-α*(hz/2) + 1j*(β2/2)*(ω**2)*(hz/2))
            Ey = Ey*np.exp(-α*(hz/2) + 1j*(β2/2)*(ω**2)*(hz/2))

            # Nonlinear step
            Ex = ifft(Ex);
            Ey = ifft(Ey);          
            Ex = Ex*np.exp(1j*γ*8/9*(np.abs(Ex)**2 + np.abs(Ey)**2)*(hz))
            Ey = Ey*np.exp(1j*γ*8/9*(np.abs(Ex)**2 + np.abs(Ey)**2)*(hz))

    #         Ex=Ex.*exp(1i*gamma*8/9*(abs(Ex).^2)*(hz));% + abs(Ey).^2)*(hz));
    #         Ey=Ey.*exp(1i*gamma*8/9*(abs(Ex).^2)*(hz));% + abs(Ey).^2)*(hz));

            # Second linear step
            Ex = fft(Ex);
            Ey = fft(Ey);
            Ex = Ex*np.exp(-α*(hz/2) + 1j*(β2/2)*(ω**2)*(hz/2))
            Ey = Ey*np.exp(-α*(hz/2) + 1j*(β2/2)*(ω**2)*(hz/2))
        
        Ex = Ex*np.exp(α*Nsteps*hz)
        Ey = Ey*np.exp(α*Nsteps*hz)
    
    Ex = ifft(Ex)
    Ey = ifft(Ey)
    
    return Ex, Ey
    
```

```python
Ltotal = 100
Lspan  = 50
hz = 0.1
alpha = 0.2
gamma = 0
D = 16
Fc = 193.1e12
Fs = 64e9
P0 = 1e-3
```

```python
from scipy import signal
import matplotlib.pyplot as plt
t       = np.linspace(-10, 10, 640, endpoint=False)
i, q, e = signal.gausspulse(t, fc=5, retquad=True, retenv=True)

plt.plot(t, i, t, q, t, e, '--')
```

```python
Ex = np.sqrt(P0/2)*e
Ey = np.sqrt(P0/2)*e

plt.plot(t, np.abs(Ex), t, np.abs(Ey),'--')
```

```python
Ex_out, Ey_out = manakovSSF(Ex, Ey, hz, Lspan, Ltotal, alpha, gamma, D, Fc, Fs)

plt.plot(t, np.abs(Ex_out), t, np.abs(Ey_out),'--')
```
